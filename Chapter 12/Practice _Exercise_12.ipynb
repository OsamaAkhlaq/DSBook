### Feature Engineering Using Data Aggregation on the Diabetes dataset

### Import modules

```python
import pandas as pd
```

### Loading the dataset 

```python
file_url = 'https://raw.githubusercontent.com/fenago/DSBook/main/Chapter%2008/diabetes.csv'
df = pd.read_csv(file_url)
```

### Perform data aggregation to find the maximum Insulin for each Glucose Level and the Blood Pressure using the .groupby.agg() method and save the results in a new DataFrame called df_agg

```python
df_agg = df.groupby(['Glucose', 'BloodPressure']).agg({'Insulin': 'max'}).reset_index()
df_agg.columns = ['Glucose', 'BloodPressure', 'InsulinMax']
```

### Print out the first five rows of df_agg

```python
df_agg.head()
```

### Merge the original DataFrame, df, to df_agg using a left join (how='left') on the Glucose and BloodPressure columns using the merge() method

```python
df_new = pd.merge(df, df_agg, how='left', on=['Glucose', 'BloodPressure'])
df_new.head()
```

### Create a new column called InsulinRatio by dividing Insulin by InsulinMax

```python
df_new['InsulinRatio'] = df_new['Insulin'] / df_new['InsulinMax']
df_new.head()
```

### . Perform data aggregation to find the maximum Age for each Neighborhood and Blood Pressure using the .groupby.agg() method and save the results in a new DataFrame called df_agg2

```python
df_agg2 = df.groupby(['Glucose', 'BloodPressure']).agg({'Age': 'max'}).reset_index()
```

### Rename the column of df_agg2 to Glucose, Blood Pressure, and AgeMax

```python
df_agg2.columns = ['Glucose', 'BloodPressure', 'AgeMax']
df_agg2.head()
```

### Merge the original DataFrame, df, to df_agg2 using a left join (how='left') on the Glucose and Blood Pressure columns using the merge() method

```python
df_final = pd.merge(df_new, df_agg2, how='left', on=['Glucose', 'BloodPressure'])
df_final['AgeRatio'] = df_final['Age'] / df_final['AgeMax']
```
