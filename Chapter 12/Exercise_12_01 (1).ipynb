### Importing Modules

```python
import pandas as pd
```

### Loading data

```python
abbr = pd.read_csv('https://raw.githubusercontent.com/fenago/DSBook/main/Chapter%2012/state-abbrevs.csv')
areas = pd.read_csv('https://raw.githubusercontent.com/fenago/DSBook/main/Chapter%2012/state-areas.csv')
```

### Print the number of unique values in this column 

```python
areas['state'].nunique()
```

### Perform a left join on the two DataFrames using the .merge() method 

```python
merged_df = pd.merge(abbr, areas, how='left', on='state')
merged_df.shape
```

### Print the total number of missing values from the 'area (sq. mi)'column by combining the .isna() and .sum() methods

```python
merged_df['area (sq. mi)'].isna().sum()
```

### Print the missing states using the .iloc() method,

```python
merged_df.loc[merged_df['area (sq. mi)'].isna(), 'state']
```
