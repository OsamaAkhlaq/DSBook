### Date Manipulations on a Landslides data

### Importing modules

```python
import pandas as pd
```

### Loading data

```python
file_url = 'https://raw.githubusercontent.com/fenago/DSBook/main/Chapter%2012/catalog.csv'
df = pd.read_csv(file_url)
```

### Display the first five rows using the .head() method:

```python
df.head()
```

### Display the datatypes:

```python
df.dtypes
```

### Convert the date attribute  to datetime using the pd.to_datetime() method

```python
df['date_parsed'] = pd.to_datetime(df['date'])
```

### Create a new column which will contain the year of each date

```python
df['Year_parsed'] = df['date_parsed'].dt.year
```

### Create a new column which will contain the month of each date

```python
df['Month_parsed'] = df['date_parsed'].dt.month
```

### Create a new column which will contain the day of the month  of each date

```python
df['Dom_parsed'] = df['date_parsed'].dt.day
```

### Create a new column which will contain the day of the week of each date

```python
df['Dow_parsed'] = df['date_parsed'].dt.dayofweek
```
