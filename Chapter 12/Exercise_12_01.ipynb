### Merging Datasets

### Importing Modules

```python
import pandas as pd
```

### Loading data

```python
df = pd.read_csv('https://raw.githubusercontent.com/PacktWorkshops/The-Data-Science-Workshop/master/Chapter12/Dataset/taxstats2015.csv')
```

### Display the dimensions of this DataFrame using the .shape attribute

```python
df.shape
```


### Displaying first 5 entries

```python
df.head()
```

### Display the number of unique values for the Postcode variable 

```python
df['Postcode'].nunique()
```

### Loading the second Postcode dataset

```python
postcode_df = pd.read_excel('https://github.com/PacktWorkshops/'\
               'The-Data-Science-Workshop/blob/'\
               'master/Chapter12/Dataset/'\
               'taxstats2016individual06taxablestatusstate'\
               'territorypostcodetaxableincome%20(2).xlsx?'\
               'raw=true', sheet_name='Individuals Table 6B', header=2)
```

### Print the dimensions of postcode_df using the .shape attribute

```python
postcode_df.shape
```

### Displaying first 5 entries

```python
postcode_df.head()
```

### Print the number of unique values in this column 

```python
postcode_df['Postcode'].nunique()
```

### Perform a left join on the two DataFrames using the .merge() method 

```python
merged_df = pd.merge(df, postcode_df, how='left', on='Postcode')
merged_df.shape
```

### Print the total number of missing values from the 'State/Territory1'column by combining the .isna() and .sum() methods

```python
merged_df['State/ Territory1'].isna().sum()
```

### Print the missing postcodes using the .iloc() method,

```python
merged_df.loc[merged_df['State/ Territory1'].isna(), 'Postcode']
```
