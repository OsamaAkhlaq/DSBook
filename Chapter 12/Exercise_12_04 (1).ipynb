### Feature Engineering Using Data Aggregation on the MIAMI Housing Dataset

### Import modules

```python
import pandas as pd
```

### Loading the dataset 

```python
file_url = 'https://raw.githubusercontent.com/fenago/DSBook/main/miami-housing.csv'
df = pd.read_csv(file_url)
```

### Perform data aggregation to find the maximum SalePrice for each LATITUDE and the LONGITUDE using the .groupby.agg() method and save the results in a new DataFrame called df_agg

```python
df_agg = df.groupby(['LATITUDE', 'LONGITUDE']).agg({'SALE_PRC': 'max'}).reset_index()
df_agg.columns = ['LATITUDE', 'LONGITUDE', 'SALE_PRCMax']
```

### Print out the first five rows of df_agg

```python
df_agg.head()
```

### Merge the original DataFrame, df, to df_agg using a left join (how='left') on the LONGITUDE and LATITUDE columns using the merge() method

```python
df_new = pd.merge(df, df_agg, how='left', on=['LATITUDE', 'LONGITUDE'])
df_new.head()
```

### Create a new column called SalePriceRatio by dividing SalePrice by SalePriceMax

```python
df_new['SalePriceRatio'] = df_new['SALE_PRC'] / df_new['SALE_PRCMax']
df_new.head()
```

### . Perform data aggregation to find the maximum LotArea for each LATITUDE and LONGITUDE using the .groupby.agg() method and save the results in a new DataFrame called df_agg2

```python
df_agg2 = df.groupby(['LATITUDE', 'LONGITUDE']).agg({'LND_SQFOOT': 'max'}).reset_index()
```

### Rename the column of df_agg2 to Longitude, Latitude, and LND_SQFOOTMax

```python
df_agg2.columns = ['LATITUDE', 'LONGITUDE', 'LND_SQFOOTMax']
df_agg2.head()
```

### Merge the original DataFrame, df, to df_agg2 using a left join (how='left') on the Longitude and Latitude columns using the merge() method

```python
df_final = pd.merge(df_new, df_agg2, how='left', on=['LATITUDE', 'LONGITUDE'])
df_final['LND_SQFOOTRatio'] = df_final['LND_SQFOOT'] / df_final['LND_SQFOOTMax']
```
