```python
#@title Importing modules
from sklearn import datasets, svm, model_selection
import pandas as pd
```

```python
#@title Loading data
# dataset
df = pd.read_csv('https://raw.githubusercontent.com/fenago/DSBook/main/Chapter%2008/diabetes.csv')

#Let's create numpy arrays for features and target
X = df.drop('Outcome',axis=1).values
y = df['Outcome'].values
```

```python
#@title support vector machine classifier

clr = svm.SVC(gamma='scale')
```

```python
#@title hyperparameter grid. contains linear and polynomial kernels

grid = [
  {'kernel': ['linear']},
  {'kernel': ['poly'], 'degree': [2, 3, 4]}
 ]
```

```python
#@title setting up the grid search to score on accuracy and evaluate over 10 folds

cv_spec = model_selection.GridSearchCV(estimator=clr, param_grid=grid, scoring='accuracy', cv=10)
```

```python
#@title Starting the grid search

cv_spec.fit(X, y)
```

```python
#@title the available information

print(cv_spec.cv_results_.keys())
```

```python
cv_spec.cv_results_
```

```python
import pandas as pd

# convert the dictionary of results to a pandas dataframe
results = pd.DataFrame(cv_spec.cv_results_)

print(
# show the hyperparameterizations
results.loc[:,['params','mean_test_score']].sort_values('mean_test_score', ascending=False)
)
```
