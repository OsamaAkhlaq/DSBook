```python
#@title Importing modules
from sklearn import neighbors, datasets, model_selection
import pandas as pd
```

```python
#@title Loading data
# dataset
df = pd.read_csv('https://raw.githubusercontent.com/fenago/DSBook/main/Chapter%2008/diabetes.csv')

#Let's create numpy arrays for features and target
X = df.drop('Outcome',axis=1).values
y = df['Outcome'].values
```

```python
#@title Initialize a k-NN classifier with its default hyperparameterization:

# no arguments specified
knn = neighbors.KNeighborsClassifier()
```

```python
#@title Feed this classifier into a 5-fold cross-validation (cv), calculating the precision score for each fold.
# 5 folds, scored on precision
cv = model_selection.cross_val_score(knn, X, y, cv=5, scoring='precision')
```

```python
#@title Printing cv shows the precision score calculated for each fold

# precision scores
print(cv)
```

```python
#@title Calculate and print the mean precision score for all folds

# average over all folds
print(round(cv.mean(), 2))
```

```python
#@title Run everything again, this time setting hyperparameter k to 14
# k = 14
knn = neighbors.KNeighborsClassifier(n_neighbors=14)

cv = model_selection.cross_val_score(knn, X, y, cv=10, scoring='precision')

print(round(cv.mean(), 2))
```

```python
#@title Try again with k = 7, 3, and 1
def evaluate_knn(k):
    knn = neighbors.KNeighborsClassifier(n_neighbors=k)
    cv = model_selection.cross_val_score(knn, X, y, cv=10, scoring='precision')
    print(round(cv.mean(), 2))

evaluate_knn(k=7)
evaluate_knn(k=3)
evaluate_knn(k=1)
```

```python
#@title Let's alter a second hyperparameter. Set k = 5
# k =5, weights evaluated using distance
knn = neighbors.KNeighborsClassifier(n_neighbors=14, weights='distance')

cv = model_selection.cross_val_score(knn, X, y, cv=10, scoring='precision')

print(round(cv.mean(), 2))
```
