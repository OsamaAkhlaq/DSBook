```python
#@title Importing data
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
#@title Loading data
data = pd.read_csv('https://raw.githubusercontent.com/fenago/DSBook/main/Chapter%2008/data.csv')
```

```python
#@title Extracting features and labels
#Get Target data 
y = data['diagnosis']

#Load X Variables into a Pandas Dataframe with columns 
X = data.drop(['id','diagnosis','Unnamed: 32'], axis = 1)
```

```python
#@title Check X variables 
X.isnull().sum()
#We do not have any missing values
```

```python
#@title First 5 entries
X.head()
```

```python
#@title Build Logistic Regression Model with hyperparameter
logModel = LogisticRegression()
param_grid = [    
    {'penalty' : ['l1', 'l2', 'elasticnet', 'none'],
    'C' : np.logspace(-4, 4, 20),
    'solver' : ['lbfgs','newton-cg','liblinear','sag','saga'],
    'max_iter' : [100, 1000,2500, 5000]
    }
]
```

```python
#@title Fit the model
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(logModel, param_grid = param_grid, cv = 3, verbose=True, n_jobs=-1)

best_clf = clf.fit(X,y)
```

```python
best_clf.best_estimator_
```

```python
#@title Check Accuracy
print (f'Accuracy - : {best_clf.score(X,y):.3f}')
```
