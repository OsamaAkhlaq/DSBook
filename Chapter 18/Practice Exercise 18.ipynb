###  Deploy a Model as a Web API

### Import modules

```python
import pandas as pd
import joblib
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

### Loading data

```python
data =  pd.read_csv('https://raw.githubusercontent.com/fenago/DSBook/main/Chapter%204/zoo2.csv')
```

### displaying first 5 entries

```python
data.head()
```

### Replace all missing values with 0 using the .fillna() method

```python
data.fillna(0, inplace=True)
```

###  Extract the 'class_type' response variable using the .pop() method:

```python

y = data.pop('class_type')
```

### Remove the animal namecolumn using the.drop() method 

```python
X = data.drop('animal_name', axis=1)
```

```python
X.head()
```

### Split into training and test sets using the train_test_split function with the parameters test_size=0.33 and random_state=888

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=888)
```

### Instantiate a RandomForestClassifier with random_state=1 and save it into a new variable called rf_model:

```python

rf_model = RandomForestClassifier(random_state=1)
rf_model.fit(X_train, y_train)
```

```python
rf_model.predict([X_test.iloc[0,]])
```

## Save the RandomForest model as a separate file called model.pkl using joblib.dump():

```python
joblib.dump(rf_model, "model.pkl")
```

### Import the socket, threading, requests, json, and numpy packages and the Flask class, as well as the jsonify and request functions 

```python
import socket
import threading
import requests
import json
from flask import Flask, jsonify, request
import numpy as np
```

```python
ip_address = socket.gethostbyname(socket.gethostname())
ip_address
```

```python
app = Flask(__name__)
```

```python
trained_model = joblib.load("model.pkl")
```

```python
@app.route('/api', methods=['POST'])
def predict():
  data = request.get_json()
  prediction = trained_model.predict(data)
  str_pred = np.array2string(prediction)
  return jsonify(str_pred)
```

```python
flask_thread = threading.Thread(target=app.run, kwargs={'host':'0.0.0.0','port':80})
flask_thread.start()
```

```python
record = X_test.iloc[0,].to_list()
record
```

```python
j_data = json.dumps([record])
```

```python
headers = {'content-type': 'application/json', 'Accept-Charset': 'UTF-8'}
```

```python
r = requests.post(f"http://{ip_address}/api", data=j_data, 
headers=headers)
r.text
```
