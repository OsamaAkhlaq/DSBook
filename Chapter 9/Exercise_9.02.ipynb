```python
#@title Importing modules
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import altair as alt
```

```python
#@title Loading data
data = pd.read_csv('https://raw.githubusercontent.com/fenago/DSBook/main/Chapter%209/tax.csv')
data.head()
```

```python
#@title Extract the Count salary and wages column using .pop() and save it into a variable called y
y = data.pop('Count salary and wages')
```

```python
#@title Print the summary of the DataFrame using .describe().
data.describe()
```

```python
#@title Split the DataFrame into training and testing sets using train_test_split() with test_size=0.3 and random_state = 1:
X_train, X_test, y_train, y_test = train_test_split(data, y, test_size=0.3, random_state=1)
```

```python
#@title Instantiate RandomForestRegressor with random_state=1, n_estimators=50, max_depth=6, and min_samples_leaf=60
rf_model = RandomForestRegressor(random_state=1, \
                                 n_estimators=50, max_depth=6,\
min_samples_leaf=60)
```

```python
#@title Train the model on the training set using .fit()
rf_model.fit(X_train, y_train)
```

```python
#@title Predict the outcomes of the training and testing sets using .predict():
preds_train = rf_model.predict(X_train)
preds_test = rf_model.predict(X_test)
```

```python
#@title Calculate the mean squared error on the training set and print its value:
train_mse = mean_squared_error(y_train, preds_train)
train_mse 
```

```python
#@title Calculate the mean squared error on the testing set and print its value:
test_mse = mean_squared_error(y_test, preds_test)
test_mse
```

```python
#@title train_mse = mean_squared_error(y_train, preds_train)
rf_model.feature_importances_
```

```python
#@title Create an empty DataFrame called varimp_df:
varimp_df = pd.DataFrame()
varimp_df['feature'] = data.columns
varimp_df['coefficient'] =  rf_model.feature_importances_
```

```python
#@title Print the first five rows of  varimp_df:
varimp_df.head()
```

```python
#@title Plot a bar chart with Altair using coef_df and coefficient as the x axis and feature as the y axis
alt.Chart(varimp_df).mark_bar().encode(
    x='coefficient',
    y="feature"
)
```
