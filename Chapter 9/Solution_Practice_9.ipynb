```python
#@title Import Modules
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.inspection import plot_partial_dependence
import altair as alt
```

```python
#@title Loading Data
df = pd.read_csv('https://raw.githubusercontent.com/PacktWorkshops/The-Data-Science-Workshop/master/Chapter09/Dataset/phpYYZ4Qc.csv')
```

```python
#@title  Extract "rej" from data
y = df.pop('rej')
```

```python
#@title Split the DataFrame into training and testing sets using train_test_split() with test_size=0.3 and random_state = 1:
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.3, random_state=1)
```

```python
#@title Instantiate RandomForestRegressor with random_state=1, n_estimators=50, max_depth=6, and min_samples_leaf=60:
rf_model = RandomForestRegressor(random_state=1, n_estimators=50, max_depth=6, min_samples_leaf=60)
```

```python
#@title Train the model on the training set using .fit():
rf_model.fit(X_train, y_train)
```

```python
#@title Plot the partial dependence plot using plot_partial_dependence()
plot_partial_dependence(rf_model, X_test, features=[df.columns.get_loc('a1pop')]) 
```
