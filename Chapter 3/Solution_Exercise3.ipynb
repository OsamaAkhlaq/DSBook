```python
import numpy as np 
import pandas as pd 

from sklearn import preprocessing
import matplotlib.pyplot as plt 
plt.rc("font", size=14)
import seaborn as sns
sns.set(style="white") #white background style for seaborn plots
sns.set(style="whitegrid", color_codes=True)

import warnings
warnings.simplefilter(action='ignore')
```

```python
# Read CSV train data file into DataFrame
train_df = pd.read_csv("https://raw.githubusercontent.com/fenago/DSBook/main/Chapter%203/gender_submission.csv")

# preview train data
train_df.head()
```

```python
print('The number of samples into the train data is {}.'.format(train_df.shape[0]))
```

### Data Quality & Missing Value Assessment

```python
# check missing values in train data
train_df.isnull().sum()
```

### Survived - Missing Values

```python
# percent of missing "Age" 
print('Percent of  "Survived" records is %.2f%%' %((train_df['Survived'].isnull().sum()/train_df.shape[0])*100))
```

### Density Plot

```python
ax = train_df["Survived"].hist(bins=15, density=True, stacked=True, color='teal', alpha=0.6)
train_df["Survived"].plot(kind='density', color='teal')
ax.set(xlabel='Survived')
plt.xlim(-10,85)
plt.show()
```
