### Finding a relationship between curtosis and the class for notes 

```python
# Relationship between skewness and authentication of note 
notes_data.groupby(['curtosis', 'class'])['class']\
        .agg(CurtosisToT='count').reset_index()
```

```python
"""
Relationship between ENTROPY and Authentication"Class"
"""
notes_data.groupby(['entropy', 'class'])['class']\
        .agg(entropyToT='count').reset_index()
```
